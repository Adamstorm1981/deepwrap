#Language Translation in dltkdl

dltkdl supports language translation. There are currently two different classes to support language translation. 

1) EnglishTranslator for translation to English and 

2) Translator for translations between many different languages. Both are wrappers around MarianMT models in the transformers library.

###The EnglishTranslator Class for Translation to English

The **EnglishTranslator** class can be used to easily convert text in various languages to English. It currently supports the following languages:

zh : Chinese (both Simplified and Traditional)

ar : Arabic

ru : Russian

de : German

af : Afrikaans

fr : French

es : Spanish

it : Italian

pt : Portuguese

To demonstrate such translations, let us first begin by importing dltkdl.

````
import deepwrap
from deepwrap import text
````

Next, we will translate the following sentence to English from a number of different source languages.



```
# We’re born alone, we live alone, we die alone. Only through our love and friendship can we create the illusion for the moment that we’re not alone.
```



To generate translations to English from a particuar language, we simply supply the language code for the language as the src_lang argument to EnglishTranslator.



###Arabic to English

````
translator = text.EnglishTranslator(src_lang='ar')
source_text = '''لقد ولدنا وحدنا ، نعيش وحدنا ، نموت وحدنا. فقط من خلال حبنا وصداقتنا يمكننا أن نخلق الوهم للحظة بأننا لسنا وحدنا.'''
print(translator.translate(source_text))
````

The translate method can accept single sentences, paragraphs, or entire documents. However, if the document is large (e.g., a book), we recommend that you break it up into smaller chunks (e.g., pages or paragraphs). This is because dltkdl tokenizes your document into individual sentences, which are joined together and fed to model as single batch when generating the translation. If the batch is too large for memory, errors will occur.

When instantiating the EnglishTranslator, pretrained models are automatically loaded, which may take a few seconds. Once instantiated, the translate method can be repeatedly invoked on different documents or sentences. 

Next, let us reinstantiate an EnglishTranslator object to translate Russian.

###Russian to English


````
translator = text.EnglishTranslator(src_lang='ru')
source_text = '''Мы рождаемся одни, мы живем одни, мы умираем одни. 
Только благодаря нашей любви и дружбе мы можем создать иллюзию на данный момент, что мы не одиноки.'''
print(translator.translate(source_text))
````

###Spanish to English

````
translator = text.EnglishTranslator(src_lang='es')
source_text = '''Nacemos solos, vivimos solos, morimos solos. 
Solo a través de nuestro amor y amistad podemos crear la ilusión por el momento de que no estamos solos.'''
print(translator.translate(source_text))
````

###Portuguese to English

````
translator = text.EnglishTranslator(src_lang='pt')
source_text = '''Nós nascemos sozinhos, vivemos sozinhos, morremos sozinhos. 
Somente através do nosso amor e amizade podemos criar a ilusão de que não estamos sozinhos.'''
print(translator.translate(source_text))
````

##The Translator Class for Translating to and from Many Languages

For translations from and to other languages, text.Translatorinstances can be used. Translator instances accept as input a pretrained model from Helsinki-NLP. 

For instance, to translate English to Dutch, one can use the Helsinki-NLP/opus-mt-en-nl model:

````
translator = text.Translator(model_name='Helsinki-NLP/opus-mt-en-nl')
source_text = '''My name is Sarah and I live in London.'''
print(translator.translate(source_text))
````
