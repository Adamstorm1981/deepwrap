deepwrap includes the ability to summarize text based on a pretrained BART model from the transformers library.

To perform summarization, first create a TransformerSummarizer instance as follows. ***(Note that this feature requires PyTorch to be installed on your system.)***

```
from deepwrap import text
text_sum = text.TransformerSummarizer()
```
